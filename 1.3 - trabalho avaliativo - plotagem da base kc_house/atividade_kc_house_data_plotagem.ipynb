## MAPA DE CIRCULOS INFORMANDO AS CASAS COM QUANTIDADE DE QUARTOS E VALOR DESEJAVEIS, VALOR DESEJAVEL MAS NÃO ATENDE A QUANTIDADE DE QUARTOS OU ATENDE A QUANTIDADE DE QUARTOS MAS SUPERA O VALOR DESEJADO

```python
import pandas as pd
import folium
import subprocess
import sys


try:
    import folium
except ImportError:
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'folium'])


df = pd.read_csv('kc_house_data.csv')


min_quartos = int(input("Informe a quantidade mínima de quartos: "))
max_preco = float(input("Informe o valor máximo disposto a pagar: "))


def filter_houses(df, min_quartos, max_preco):
    return df[(df['bedrooms'] >= min_quartos) & (df['price'] <= max_preco)]

df_filtrado = filter_houses(df, min_quartos, max_preco)


def create_map(dataset, min_quartos, max_preco):

    mapa = folium.Map(location=[dataset['lat'].mean(), dataset['long'].mean()], zoom_start=12)


    for index, row in dataset.iterrows():
        if row['bedrooms'] >= min_quartos and row['price'] <= max_preco:
            color = 'green'
        elif row['bedrooms'] >= min_quartos and row['price'] > max_preco:
            color = 'red'
        elif row['bedrooms'] < min_quartos and row['price'] <= max_preco:
            color = 'orange'
        else:
            continue

        folium.CircleMarker(
            location=[row['lat'], row['long']],
            radius=5,
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.7,
            popup=folium.Popup(f"Preço: ${row['price']:.0f}<br> Quartos: {row['bedrooms']}", max_width=300)
        ).add_to(mapa)

    return mapa


mapa_resultado = create_map(df, min_quartos, max_preco)


mapa_resultado.save('mapa_casas_filtradas.html')

print("Veja o mapa com as seguintes instruções: MARCADORES VERDES (Atende aos requisitos), MARCADORES LARANJAS (Atende o valor mas não a quantidade de quartos) e os MARCADORES VERMELHOS (Atende a quantidade de quartos mas não o valor))")
print("Mapa gerado com sucesso!")
```
